## Best associates for France:

```python
france_data = retail_data[retail_data['Country']=='France']
france_data
```

```python
france_data_pivot = pd.pivot_table(data= france_data, values='Quantity', index='InvoiceNo', columns='Description').fillna(0)
france_data_pivot
```

```python
france_encoded_data = france_data_pivot.applymap(lambda x:1 if x>0 else 0)
france_encoded_data
```

## 4. Data Preparation

```python
pd.pivot_table(data = uk_data, values='Quantity', index='InvoiceNo', columns='Description').fillna(0)
```

## 5. Model Building:

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
apriori()
```

```python
association_rules()
```

```python
freq_itemsets = apriori(df = france_encoded_data, min_support = 0.02, use_colnames=True)
freq_itemsets
```

```python
freq_itemsets.to_csv('Frequent_itemsets.csv')
```

```python
best_associates = association_rules(df = freq_itemsets, metric='confidence', min_threshold=0.70)
best_associates
```

```python
best_associates.to_csv('Best_associates.csv')
```
