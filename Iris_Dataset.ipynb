```python
x =iris_df.iloc[:,:3].values
```

## Fitting:

### K-Means:

```python
from sklearn.cluster import KMeans
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', n_init=10, max_iter=300, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
#Plotting

plt.plot(range(1,11), wcss)
plt.title('Elbow Curve')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#The elbow point here is 3. We can say that no. of optimal clusters here is 3.
```

#### Now that we have optimum number of clusters, n=3, we can build the model.

```python
#Build Cluster Algorithm
new_cluster = KMeans(n_clusters=3, random_state=0)
y_k=new_cluster.fit_predict(x)
```
